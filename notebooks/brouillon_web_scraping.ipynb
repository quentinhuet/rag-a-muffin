```python
import requests
import bs4
```

### Scrapping de la page de recherche Marmiton pour en tirer les URLs des recettes

On note que l'on peut accéder à la page de recherche Marmiton, sur la recherche des muffins, via l'url suivante : https://www.marmiton.org/recettes/recherche.aspx?aqt=muffin&page=1

Le terme "page=1" permet de choisir la page de recherche. Ainsi, on peut itérer sur les pags de recherche, tant qu'elles existent, pour accéder à toutes les pages HTML de recherche de recettes de muffins. 

```python
response = requests.get('https://www.marmiton.org/recettes/recherche.aspx?aqt=muffin&page=3')
try:
    response.raise_for_status()
except Exception as exc:
    print(f'There was a problem: {exc}')
```

On simplifie la lecture du document avec le module `beautifulsoup4`.

```python
page_soup = bs4.BeautifulSoup(response.text, 'html.parser')
```

```python
resultats = []
nb_resultats = 0
page = 1

while True: 

    response = requests.get(f'https://www.marmiton.org/recettes/recherche.aspx?aqt=muffin&page={page}')
    try:
        response.raise_for_status()
    except Exception as exc:
        print(f'There was a problem: {exc}')
        break

    page_soup = bs4.BeautifulSoup(response.text, 'html.parser')

    for lien in page_soup.find_all('a'):
        url_partielle = lien.get('href')
        
        if url_partielle and '/recettes/recette_' in url_partielle:
            
            if url_partielle.startswith('/'):
                url_complete = "https://www.marmiton.org" + url_partielle
            else:
                url_complete = url_partielle
                
            titre_brut = lien.text 
            titre_propre = titre_brut.strip() 
            
            if "muffin" in titre_propre.lower(): #on vérifie qu'on a bien que des recettes de muffins sur la page.
                resultats.append({
                    "titre": titre_propre,
                    "url": url_complete
                })

    print(f'Page de recherche {page} scrappée.')
    print (f'Nombre de résultats scrappés sur la page {page} : {len(resultats)-nb_resultats}')
    nb_resultats = len(resultats)
    page += 1
    

print(f"J'ai trouvé {len(resultats)} recettes :")
for r in resultats:
    print(f"- {r['titre']} -> {r['url']}")
```
